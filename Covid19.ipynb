```python
#импорт библиотек
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
import plotly.figure_factory as ff
data = pd.read_csv('covid19-russia-cases-scrf.csv')
data.head()
```

```python
#построение графика по суммарному количеству заболевших к определенной дате
df1 = data.groupby(['Date', 'Region/City'])['Confirmed', 'Deaths', 'Recovered'].sum().reset_index()
px.line(df1, x = 'Date', y = 'Confirmed', color = 'Region/City', title = 'Daily confirmed cases spread', height = 600)
```

```python
#построение графика по суточной заболеваемости
df2 = data.groupby(['Date', 'Region/City'])[['Day-Confirmed', 'Confirmed']].max().reset_index()
df2 = df2.sort_values(['Confirmed', 'Date'], ascending=False)

px.line(df2, x = 'Date', y = 'Day-Confirmed', color = 'Region/City', title='Daily confirmed cases spread', height=600)
```

```python
#построение круговой диаграммы 
fig = px.pie(df1, values='Confirmed', names='Region/City', title='Number of new cases')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
